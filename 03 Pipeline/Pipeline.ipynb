# Step 3 - Separate x and y
- drop id column

```python
X = df.drop(columns=["id", "Weight"])
X.head()
Y = df[["Weight"]]
```

```python
Y
```

# Step 4 - Apply preprocessing

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

#### Pipeline


```python
from sklearn.pipeline import make_pipeline  # used to apply multiple preprocessing steps
from sklearn.impute import SimpleImputer  # used to replace missing values
from sklearn.preprocessing import (
    StandardScaler,
    OneHotEncoder,
)  # Preprocessing on numerical and categorical data
from sklearn.compose import (
    ColumnTransformer,
)  # To combine categorical and numerical pipeline
```

```python
num_pipe = make_pipeline(SimpleImputer(strategy="median"), StandardScaler())
```

```python
cat_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop="first"),
)
```

```python
pre = ColumnTransformer(
    [("num", num_pipe, con), ("cat", cat_pipeline, cat)]
).set_output(transform="pandas")
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

# Step 5 - Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    X_pre, Y, test_size=0.2, random_state=21
)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

# Step 6 - Model building

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
model.intercept_
```

```python
model.coef_
```

# Step 7 - Model evaluation

```python
# Train r2 score
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score,
)
```

```python
def evaluatemodel(model, x, y):
    ypred = model.predict(x)

    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)

    print(f"rmse: {rmse:.2f}")
    print(f"mae: {mae:.2f}")
    print(f"mape: {mape:.2%}")
    print(f"r2: {r2:.2%}")
```

```python
evaluatemodel(model, xtrain, ytrain)
```

```python
evaluatemodel(model, xtest, ytest)
```
